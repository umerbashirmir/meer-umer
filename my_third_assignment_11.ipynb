````python
# Question 2 : Explain the concept of function arguments and parameters in Python .
Answer : In Python, **function parameters** and **arguments** refer to the way values are passed into a function when it is called.
### 1. **Function Parameters**:
   - Parameters are the variables listed inside the function definition. They define what kind of input the function expects.
   - For example, in the function `def add(x, y):`, `x` and `y` are parameters.

### 2. **Function Arguments**:
   - Arguments are the actual values passed to the function when it is called.
   - For example, in the function call `add(3, 5)`, `3` and `5` are arguments.

### Example:

def add(x, y):  # x and y are parameters
    return x + y

result = add(3, 5)  # 3 and 5 are arguments
print(result)  # Output: 8
```

### Key Points:
- **Parameters**: Define the input the function expects when it's written (e.g., `x` and `y` in the `add` function).
- **Arguments**: The actual values provided when the function is called (e.g., `3` and `5` in `add(3, 5)`).

### Types of Arguments in Python:
1. **Positional Arguments**: Passed in the order in which parameters are defined.
2. **Keyword Arguments**: Passed by explicitly naming the parameter (e.g., `add(x=3, y=5)`).
3. **Default Arguments**: Parameters that have default values if no argument is provided.
4. **Variable-length Arguments**: When you want to accept an arbitrary number of arguments (using `*args` and `**kwargs`).

Example with default and keyword arguments:

def greet(name, message="Hello"):  # 'message' has a default value
    return f"{message}, {name}!"

print(greet("Alice"))  # Uses default message -> "Hello, Alice!"
print(greet("Bob", "Good morning"))  # Uses provided message -> "Good morning, Bob!"
````

````python
# Question 3 : What are the different ways to define and call a function in Python?
Answer : In Python, functions can be defined and called in several different ways. Below are the main ways to define and call functions, with examples:

### 1. **Defining and Calling a Regular Function**

A regular function is defined using the `def` keyword followed by the function name and parameters.
You call it by its name with arguments passed inside parentheses.

#### Example:
# Function definition
def greet(name):
    return f"Hello, {name}!"

# Function call
result = greet("Alice")
print(result)  # Output: Hello, Alice!
```

### 2. **Defining and Calling a Function with Default Arguments**

You can provide default values for function parameters. If no argument is passed during the function call, the default value will be used.

#### Example:
# Function definition with a default argument
def greet(name, message="Hello"):
    return f"{message}, {name}!"

# Function call with a default message
result1 = greet("Alice")
print(result1)  # Output: Hello, Alice!

# Function call with a custom message
result2 = greet("Bob", "Good morning")
print(result2)  # Output: Good morning, Bob!
```

### 3. **Defining and Calling a Function with Variable-Length Arguments (Using `*args` and `**kwargs`)**

- `*args` allows the function to accept a variable number of positional arguments.
- `**kwargs` allows the function to accept a variable number of keyword arguments.

#### Example with `*args`:
# Function definition with *args
def sum_numbers(*args):
    return sum(args)

# Function call with multiple arguments
result = sum_numbers(1, 2, 3, 4)
print(result)  # Output: 10
```

#### Example with `**kwargs`:
# Function definition with **kwargs
def display_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

# Function call with keyword arguments
display_info(name="Alice", age=25)
# Output:
# name: Alice
# age: 25
```

### 4. **Lambda Functions (Anonymous Functions)**

A **lambda function** is a small anonymous function defined with the `lambda` keyword. It's typically used for short, simple operations.

#### Example:
# Lambda function definition
add = lambda x, y: x + y

# Function call
result = add(5, 3)
print(result)  # Output: 8
```

### 5. **Function Call with Return Value**

A function can return a value, which can be used directly or assigned to a variable.

#### Example:
# Function definition with return value
def multiply(x, y):
    return x * y

# Function call and using the returned value
result = multiply(4, 5)
print(result)  # Output: 20

### 6. **Function Call Using Function as an Argument**

You can pass one function as an argument to another function.

#### Example:
# Function definition that accepts another function as an argument
def apply_function(func, value):
    return func(value)

# Function to square a number
def square(x):
    return x * x

# Function call
result = apply_function(square, 6)
print(result)  # Output: 36
```

### Summary of Ways to Define and Call a Function in Python:
1. **Regular Function**: `def func():` and call it as `func()`.
2. **With Default Arguments**: `def func(name, message="Hello")` and call it as `func(name)`.
3. **With Variable-Length Arguments**: Use `*args` or `**kwargs` in the function definition.
4. **Lambda Function**: An anonymous function defined with `lambda`.
5. **Return Values**: Functions that return a value can be used in expressions.
6. **Passing Functions as Arguments**: Functions can accept other functions as parameters.

Each of these allows for flexibility in how functions are used and
 provides various tools for different types of problems.
````

````python
# Question 4 : What is the purpose of the `return` statement in a Python function?
Answer : The `return` statement in a Python function is used to **exit** the function and **send a result back** to the caller.
It allows a function to return a value (or multiple values) after performing its task, making the result accessible outside of the function.

### Purpose of `return`:
1. **Exit the Function**: When the `return` statement is executed, the function terminates, and no further code in the function is executed.
2. **Return a Value**: The value provided after the `return` keyword is sent back to the caller.
 This value can then be used for further operations or displayed.

### Syntax:
return <value>
```

If no value is provided after `return`, the function returns `None` by default.

### Examples:

#### 1. **Returning a Value**:
In this example, the function returns the result of adding two numbers:

def add(x, y):
    return x + y

result = add(5, 3)
print(result)  # Output: 8

- In this case, the function `add()` calculates the sum of `x` and `y`, then returns the result (`8`), which is stored in the variable `result` and printed.

#### 2. **Returning Multiple Values**:
You can return multiple values from a function by separating them with commas. Python automatically packs these into a tuple.

def get_full_name(first_name, last_name):
    return first_name, last_name

full_name = get_full_name("John", "Doe")
print(full_name)  # Output: ('John', 'Doe')

- The function `get_full_name()` returns two values: `first_name` and `last_name`. When called, it returns them as a tuple (`('John', 'Doe')`), which is assigned to the variable `full_name`.

#### 3. **Returning None (Implicitly)**:
If a function does not explicitly return a value, it returns `None` by default.
def greet(name):
    print(f"Hello, {name}!")

result = greet("Alice")
print(result)  # Output: None

- The function `greet()` does not have a `return` statement, so it returns `None` by default. This value is stored in the variable `result`, and printing it shows `None`.

#### 4. **Exiting a Function Early**:
The `return` statement can also be used to exit the function early before reaching the end of the function body.

def check_even_or_odd(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

result = check_even_or_odd(4)
print(result)  # Output: Even

- The `return` statement exits the function as soon as a result is determined (whether the number is even or odd),
so the function doesn't continue to execute any further code after the return.

#### 5. **Returning `None` Explicitly**:
If you use `return` without any expression, the function will return `None`.

def say_goodbye():
    return  # Implicitly returns None

result = say_goodbye()
print(result)  # Output: None
```

- The function `say_goodbye()` uses `return` without an expression, which causes it to return `None`.

### Key Points:
- The `return` statement allows a function to provide output (values) to the calling code.
- If no `return` is used, or if `return` is used without an expression, the function returns `None`.
- Once a `return` statement is executed, the function ends immediately, and no further code in that function will be executed.

The `return` statement is essential for producing results from functions, especially when a function performs
a calculation or operation and needs to send that result back for further use.
````

````python
# Question 5 :  What are iterators in Python and how do they differ from iterables?
Answer : In Python, the terms **iterators** and **iterables** are closely related but refer to different concepts. Let's break down both:

### 1. **Iterable**:
An **iterable** is any object in Python that can return an iterator. It is an object that **can be looped over
** (iterated) using a `for` loop, or it can be passed to functions like `iter()` to obtain an iterator.

- **Examples of iterables**: Lists, tuples, strings, dictionaries, sets, and files are all iterables.
- An **iterable** implements the `__iter__()` method or `__iter__()` function to return an iterator.

#### Example of an Iterable:
# A list is an iterable
my_list = [1, 2, 3]

# You can loop over the iterable
for item in my_list:
    print(item)

- The list `my_list` is an iterable. It can be used in a `for` loop or passed to `iter()` to create an iterator.

### 2. **Iterator**:
An **iterator** is an object that represents a stream of data. It keeps track of the state as you loop through
the data and knows how to get the next item in the sequence. An iterator must implement two methods:
- `__iter__()` – This method returns the iterator object itself.
- `__next__()` – This method returns the next item in the sequence. If there are no more items, it raises the
 `StopIteration` exception to signal the end of iteration.

Once an iterable is passed to `iter()`, an iterator is created. The iterator can be used to loop
through the iterable using the `next()` function or inside a `for` loop.

#### Example of an Iterator:
# Creating an iterator from an iterable
my_list = [1, 2, 3]
my_iterator = iter(my_list)  # This converts the iterable into an iterator

# Using next() to get items from the iterator
print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
print(next(my_iterator))  # Output: 3
# The next call will raise StopIteration
# print(next(my_iterator))  # Raises StopIteration
```

- The `iter(my_list)` call converts the list into an iterator, and `next(my_iterator)` returns one item at a time from the list.

### Key Differences Between Iterables and Iterators:

| **Feature**              | **Iterable**                                      | **Iterator**                                         |
|--------------------------|---------------------------------------------------|------------------------------------------------------|
| **Definition**            | An object that can be iterated over.              | An object that keeps track of the current iteration.  |
| **Methods**               | Implements `__iter__()` method.                   | Implements both `__iter__()` and `__next__()` methods.|
| **Use**                    | Can be used in a `for` loop or passed to `iter()`. | Used to iterate through an iterable's elements.      |
| **State**                  | Does not maintain state for iteration.            | Maintains the state and remembers the current position in iteration. |
| **Example**               | Lists, Tuples, Strings, Sets, etc.                | An iterator created by `iter()` from an iterable.    |

### Example: Iterables and Iterators Together
# List (Iterable)
numbers = [1, 2, 3, 4, 5]

# Create an iterator from the iterable
numbers_iterator = iter(numbers)

# Using the iterator to iterate through the list
print(next(numbers_iterator))  # Output: 1
print(next(numbers_iterator))  # Output: 2
print(next(numbers_iterator))  # Output: 3
```

### Important Points:
- An **iterable** is an object that can return an iterator (e.g., a list or a string).
- An **iterator** is an object that keeps track of where we are in the iteration and provides the next value when asked.
- You can obtain an iterator from an iterable using the `iter()` function. Once you have an iterator, you can use `next()`
 to retrieve items one at a time.
- Once an iterator is exhausted (i.e., there are no more items to return), it raises the `StopIteration` exception.

### Summary:
- **Iterable**: An object that can be iterated over (e.g., list, string, set). It provides an iterator when you call `iter()` on it.
- **Iterator**: An object that keeps track of the iteration and returns the next item with `next()`. It must implement
 both `__iter__()` and `__next__()`.

Iterables provide the sequence, while iterators allow you to traverse and access each element in the sequence.
````

````python
# Question 6 :  Explain the concept of generators in Python and how they are defined.
Answer : **Generators in Python**
A **generator** in Python is a type of iterable, like a list or a tuple, but unlike lists or tuples, **generators are lazy**,
 meaning they do not store all the values in memory at once. Instead, they generate values one at a time as you iterate over them.
  This makes generators more memory efficient, especially when dealing with large datasets or sequences.

Generators are defined using either:
1. A function with the `yield` keyword, or
2. Generator expressions (a concise, inline version).

### **Key Characteristics of Generators**:
1. **Lazy Evaluation**: They produce items one at a time when requested, not all at once.
2. **Memory Efficiency**: Since they don’t store all the values in memory, they are more memory-efficient than lists,
especially for large datasets.
3. **Stateful**: Generators remember the state of their execution. When you call `next()` on a generator, it picks up where it left off.
4. **Can be iterated only once**: Once a generator has been exhausted, it cannot be reused unless recreated.

### **1. Defining a Generator with the `yield` Keyword**

The most common way to define a generator is by using the `yield` keyword in a function. When `yield` is used,
the function becomes a generator function and returns a generator object.

- The `yield` statement produces a value and **pauses** the function’s execution. When `next()` is called again on the generator,
it resumes execution from where it left off.

#### Example 1: Basic Generator Using `yield`

# Generator function to yield numbers from 1 to 3
def count_up_to_three():
    yield 1
    yield 2
    yield 3

# Creating a generator object
gen = count_up_to_three()

# Iterating through the generator using a for loop
for num in gen:
    print(num)

**Output**:
```
1
2
3

- The function `count_up_to_three()` is a generator because it uses `yield` to return values.
- When you iterate over the generator (`gen`), it yields `1`, then `2`, then `3`.

#### Example 2: Generator with `yield` and State

# Generator function to yield squares of numbers
def generate_squares(n):
    for i in range(1, n+1):
        yield i * i

# Create a generator object for squares of numbers 1 to 5
square_gen = generate_squares(5)

# Iterating over the generator
for square in square_gen:
    print(square)

**Output**:
```
1
4
9
16
25
```

- The generator `generate_squares()` yields the square of each number from `1` to `n`.
- It calculates each square lazily, i.e., only when requested, and does not store the entire list of squares in memory.

### **2. Generator Expressions**

Generator expressions are similar to list comprehensions, but instead of creating a full list, they return a generator object.
They are written using parentheses `()` instead of square brackets `[]`.

#### Example 3: Generator Expression

# Generator expression to generate squares of numbers from 1 to 5
gen_exp = (x * x for x in range(1, 6))

# Iterating through the generator expression
for square in gen_exp:
    print(square)
```

**Output**:
```
1
4
9
16
25
```

- The generator expression `(x * x for x in range(1, 6))` is a compact way to create a generator
that yields the squares of numbers from `1` to `5`.
- As with other generators, the values are generated one at a time as you iterate through the expression.

### **3. Using `next()` with Generators**

You can use the `next()` function to manually get the next value from a generator. Each call to `next()`
 continues from where the generator left off.

#### Example 4: Using `next()` on a Generator

# A generator that yields numbers from 1 to 3
def simple_gen():
    yield 1
    yield 2
    yield 3

# Creating a generator object
gen = simple_gen()

# Using next() to get values one by one
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3

# The next call will raise StopIteration, as the generator is exhausted
# print(next(gen))  # Raises StopIteration
```

- When `next(gen)` is called, the generator produces the next value (`1`, then `2`, then `3`).
- Once the generator is exhausted, calling `next()` again raises a `StopIteration` exception.

### **4. Benefits of Generators**
- **Memory Efficiency**: Generators do not store the entire list of values in memory at once.
- **Performance**: For large datasets, generators allow you to process items one at a time, improving performance and reducing memory usage.
- **Lazy Evaluation**: Items are only computed when needed, which is useful when you are working with large or infinite sequences.

### **5. Example: Infinite Sequence Generator**

Here’s an example of an infinite sequence generator that will continue producing numbers indefinitely:

# Infinite generator to produce numbers starting from 1
def infinite_count():
    num = 1
    while True:
        yield num
        num += 1

# Creating an infinite generator
gen = infinite_count()

# Get the first 5 numbers from the infinite generator
for _ in range(5):
    print(next(gen))
```

**Output**:
1
2
3
4
5

- The generator `infinite_count()` continues indefinitely. It will keep generating numbers starting from `1`,
but we limit the iteration to 5 numbers using a `for` loop.

### **Summary**:

- **Generators** are functions that use the `yield` keyword to produce a sequence of values lazily.
- They allow efficient memory usage because they don’t store all values in memory at once.
- **Generator Expressions** provide a shorthand way to create generators using syntax similar to list comprehensions.
- **Advantages** of generators include lazy evaluation, memory efficiency, and the ability to represent infinite sequences.

Generators are especially useful when working with large data sets or when you need to represent infinite
 sequences without consuming a lot of memory.
````

````python
# Question 7 :  What are the advantages of using generators over regular functions?
Answer : Generators offer several advantages over regular functions, particularly when dealing with large datasets, memory management,
and performance. Below are the primary advantages of using generators, along with examples to illustrate how they differ from regular functions.

### **1. Memory Efficiency**
Regular functions return the entire result at once, storing all values in memory. In contrast, **generators yield values one at a time**,
 which makes them more memory-efficient, especially for large datasets or infinite sequences.

#### Example 1: Memory Usage with Regular Function vs. Generator

Let's compare a regular function that returns a list with a generator that yields values one by one.

**Using a regular function** (storing the entire result in memory):

# Regular function returning a list
def get_squares(n):
    return [x * x for x in range(n)]

# Call the function and store the result
squares = get_squares(1000000)  # A list with 1 million elements
print(len(squares))  # Output: 1000000
```
- **Memory Usage**: The entire list of 1 million squares is stored in memory, which could consume significant memory resources.

**Using a generator** (yielding values one by one):

# Generator function to yield squares one by one
def generate_squares(n):
    for x in range(n):
        yield x * x

# Creating a generator object
square_gen = generate_squares(1000000)

# Iterating over the generator without storing all values in memory
count = 0
for square in square_gen:
    count += 1
print(count)  # Output: 1000000
```
- **Memory Usage**: The generator does not store the entire sequence of squares in memory.
It generates each value on-demand as you iterate over it. This makes it far more memory-efficient, especially for large `n`.

### **2. Performance (Lazy Evaluation)**
Generators support **lazy evaluation**, meaning that they compute values only when needed,
 rather than all at once. This can improve performance, especially when you
 don't need to iterate through the entire sequence or if you're working with an infinite sequence.

#### Example 2: Lazy Evaluation with Regular Function vs. Generator

**Using a regular function** (eager evaluation):

# Regular function returning a list
def get_numbers():
    print("Generating numbers...")
    return [x for x in range(1, 6)]

# Call the function
numbers = get_numbers()
print(numbers)
```
- **Evaluation**: The function eagerly evaluates and stores the entire list in memory before it can be used.
 This can be inefficient if you only need a few values.

**Using a generator** (lazy evaluation):

# Generator function yielding values one by one
def generate_numbers():
    print("Generating numbers...")
    for x in range(1, 6):
        yield x

# Create a generator object
number_gen = generate_numbers()

# Iterating over the generator (values are generated lazily)
for num in number_gen:
    print(num)
```
- **Evaluation**: The generator only computes the next value in the sequence when needed (on each iteration),
and the "Generating numbers..." message is printed only once at the beginning.

### **3. Infinite Sequences**
Generators can be used to create **infinite sequences**. Regular functions, by contrast, must return all values at once,
 which makes handling infinite sequences impossible.

#### Example 3: Infinite Sequence with a Generator

**Using a generator** to generate an infinite sequence:
# Generator for an infinite sequence of numbers
def infinite_count():
    num = 1
    while True:
        yield num
        num += 1

# Create an infinite generator
gen = infinite_count()

# Get the first 5 numbers from the infinite sequence
for _ in range(5):
    print(next(gen))  # Output: 1, 2, 3, 4, 5
```
- **Infinite Sequence**: The generator `infinite_count()` generates numbers indefinitely without ever running
 out of memory or resources. You can get as many numbers as you need, but you don’t store the entire infinite sequence in memory.

**Using a regular function** would not be feasible for infinite sequences since it requires returning a complete
 list of values upfront, which is impossible for an infinite sequence.

### **4. Improved Performance for Large Datasets**
Generators can improve performance when processing large datasets because they produce values one at a time
 and process them immediately, rather than waiting to process everything all at once.

#### Example 4: Processing Large Datasets

Suppose you need to process a large dataset (e.g., a file with a million lines) and perform some operations on each line:

**Using a regular function** (storing the entire dataset in memory):

# Regular function reading a large file into memory
def read_large_file(file_name):
    with open(file_name, 'r') as file:
        return file.readlines()  # Reads all lines into memory

# Process the entire file (would be memory intensive for large files)
lines = read_large_file('large_file.txt')
for line in lines:
    # Process each line (e.g., clean or parse the data)
    pass
```
- **Memory Usage**: The entire file is read into memory, which can be inefficient if the file is very large.

**Using a generator** (reading the file line by line):

# Generator function to read the file line by line
def read_file_line_by_line(file_name):
    with open(file_name, 'r') as file:
        for line in file:
            yield line  # Yield one line at a time

# Process the file line by line (memory efficient)
file_gen = read_file_line_by_line('large_file.txt')
for line in file_gen:
    # Process each line as it is read (memory efficient)
    pass
- **Memory Usage**: The generator reads the file line by line, keeping only the current line in memory.
This allows for processing large files without consuming large amounts of memory.

### **5. Cleaner Code with Generators**
Generators often lead to more **concise** and **clean** code, especially for iterating over sequences or when implementing custom iteration logic.

#### Example 5: Cleaner Code with Generators

Let's say you want to filter even numbers from a large list:

**Using a regular function**:
# Regular function returning a filtered list
def filter_even_numbers(numbers):
    return [num for num in numbers if num % 2 == 0]

# Call the function
even_numbers = filter_even_numbers(range(1, 10001))
print(even_numbers[:10])  # Display the first 10 even numbers
```

**Using a generator**:
# Generator function to yield even numbers
def generate_even_numbers(numbers):
    for num in numbers:
        if num % 2 == 0:
            yield num

# Create a generator object
even_number_gen = generate_even_numbers(range(1, 10001))

# Get the first 10 even numbers
for _ in range(10):
    print(next(even_number_gen))

- **Cleaner Code**: The generator approach allows you to filter and process numbers lazily,
 without needing to create a full list in memory.

````

````python
# Question 8 :  What is a lambda function in Python and when is it typically used?
Answer : ### **Lambda Functions in Python**

A **lambda function** in Python is a small, anonymous function defined using the `lambda`
 keyword. Unlike regular functions defined with `def`, lambda functions are typically used for short,
  simple operations that can be written in a single line of code.

The syntax for a lambda function is:
lambda arguments: expression
```

- **`lambda`**: Keyword that defines the function.
- **`arguments`**: A comma-separated list of arguments (similar to function parameters).
- **`expression`**: A single expression that the function evaluates and returns. Note that the expression
 is evaluated and returned automatically without the need for the `return` keyword.

### **Key Characteristics of Lambda Functions**:
1. **Anonymous**: Lambda functions do not require a name (though you can assign them to a variable).
2. **Single Expression**: Lambda functions can only contain a single expression; they cannot have multiple statements or a body like regular functions.
3. **Short-lived**: Lambda functions are often used for small tasks and are typically used where functions are needed temporarily.

### **When is a Lambda Function Typically Used?**

1. **For Simple Operations**: When you need a short function for a one-time use, and writing a full function would be overkill.
2. **In Higher-Order Functions**: Lambda functions are often used as arguments to higher-order functions
 (functions that take other functions as arguments), such as `map()`, `filter()`, and `sorted()`.
3. **In Data Processing**: Lambda functions are frequently used in operations like sorting, filtering,
or transforming data in functional programming.

### **Examples of Lambda Functions**

#### Example 1: Simple Lambda Function

A basic lambda function that takes two arguments and returns their sum.

# Lambda function to add two numbers
add = lambda x, y: x + y

# Using the lambda function
result = add(3, 5)
print(result)  # Output: 8
```

- This `lambda` function takes two arguments, `x` and `y`, and returns their sum (`x + y`).

#### Example 2: Lambda with `map()`

The `map()` function applies a function to all items in an input list (or any iterable). Lambda functions are often used with `map()` for concise code.

# Using lambda with map to square each number in a list
numbers = [1, 2, 3, 4, 5]
squared_numbers = map(lambda x: x ** 2, numbers)

# Convert the map object to a list and print the result
print(list(squared_numbers))  # Output: [1, 4, 9, 16, 25]
```

- The lambda function `lambda x: x ** 2` squares each number in the `numbers` list.

#### Example 3: Lambda with `filter()`

The `filter()` function is used to filter elements from an iterable based on a condition. Lambda functions are often used to define the condition.

# Using lambda with filter to get even numbers from a list
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = filter(lambda x: x % 2 == 0, numbers)

# Convert the filter object to a list and print the result
print(list(even_numbers))  # Output: [2, 4, 6]
```

- The lambda function `lambda x: x % 2 == 0` checks if each number is even. The `filter()` function uses this condition to filter out odd numbers.

#### Example 4: Lambda with `sorted()`

The `sorted()` function is used to sort a list, and you can pass a lambda function to specify custom sorting logic.

# List of tuples to sort by the second element
tuples = [(1, 'apple'), (3, 'orange'), (2, 'banana')]
sorted_tuples = sorted(tuples, key=lambda x: x[1])

# Print the sorted list of tuples
print(sorted_tuples)  # Output: [(1, 'apple'), (2, 'banana'), (3, 'orange')]

- The lambda function `lambda x: x[1]` sorts the list of tuples based on the second element (the fruit name).

#### Example 5: Lambda for Inline Operations

Lambda functions are often used inline, where defining a full function would be unnecessary.

# Using lambda inline to perform a quick operation
result = (lambda x, y: x * y)(5, 3)
print(result)  # Output: 15
```

- Here, the lambda function `lambda x, y: x * y` is immediately called with arguments `5` and `3`, resulting in `15`.

### **Advantages of Lambda Functions**
1. **Concise**: They allow you to write short, simple functions in one line, which can make code more readable in certain situations.
2. **Anonymous**: Since they don't require a name, lambda functions can be used as one-time, temporary functions in expressions.
3. **Functional Programming**: Lambda functions are commonly used in functional programming paradigms
 where functions are passed as arguments to higher-order functions.

### **Disadvantages of Lambda Functions**
1. **Limited to Single Expression**: Unlike regular functions, lambda functions can only
 contain a single expression, which limits their functionality.
2. **Reduced Readability**: While lambda functions are concise, they can reduce readability
 when overused or when the logic inside the lambda is complex.
3. **No Statements or Multiple Expressions**: You can't use statements (like `if`, `for`, `while`, or multiple expressions)
in lambda functions, which limits their usage.

### **Summary**

- **Lambda functions** are anonymous, short functions defined using the `lambda` keyword.
- They are typically used for simple tasks, especially in functional programming
 with functions like `map()`, `filter()`, and `sorted()`.
- The advantage of lambda functions is that they are concise, memory-efficient, and useful in places
 where you need a simple, short function for a short period of time.

### **When to Use Lambda Functions**:
- When you need a simple function for a short task.
- When the function is used temporarily (e.g., in a list operation or filter).
- When using functional programming constructs like `map()`, `filter()`, or `reduce()`.
````

````python
# Question 10 : What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
Answer : The functions `map()`, `reduce()`, and `filter()` in Python are all part of functional programming tools that
 help process iterables (like lists, tuples, etc.). Although they are similar in some ways, they have distinct purposes and usage patterns.

### **1. `map()` Function**
- **Purpose**: `map()` applies a given function to each item of an iterable (or multiple iterables) and returns
 a new iterable (in the form of a map object, which can be converted to other collections like lists or tuples).
- **Output**: A transformed iterable where the function is applied to each element.
- **Usage**: Typically used when you want to apply the same operation to all elements of an iterable.

#### Syntax:

map(function, iterable, ...)
```

- **`function`**: The function that will be applied to each element of the iterable.
- **`iterable`**: The iterable whose elements are passed to the function.

#### Example:

# Using map to square each number
numbers = [1, 2, 3, 4]
squared_numbers = map(lambda x: x ** 2, numbers)
print(list(squared_numbers))  # Output: [1, 4, 9, 16]

### **2. `reduce()` Function**
- **Purpose**: `reduce()` is used to apply a given function cumulatively to the items of an iterable, reducing the iterable to a single result.
 It takes two elements at a time, applies the function to them, and then continues applying the function to the result and the next element.
- **Output**: A single value that results from reducing the iterable.
- **Usage**: Typically used when you want to reduce an iterable to a single accumulated value, such as summing or multiplying all elements.

#### Syntax:

from functools import reduce
reduce(function, iterable, [initial])
```

- **`function`**: The function that will be applied cumulatively to the iterable elements.
- **`iterable`**: The iterable whose elements will be processed.
- **`initial`** (optional): An initial value to start the accumulation. If not provided, the first element of the iterable is used as the initial value.

#### Example:
from functools import reduce

# Using reduce to sum all numbers
numbers = [1, 2, 3, 4]
total = reduce(lambda x, y: x + y, numbers)
print(total)  # Output: 10
```

- In this example, `reduce()` is used to accumulate the sum of the numbers, starting with the first element and
applying the `lambda x, y: x + y` function iteratively.

### **3. `filter()` Function**
- **Purpose**: `filter()` filters elements from an iterable based on a condition specified by a function.
 It only includes elements for which the function returns `True`.
- **Output**: An iterable (usually converted to a list) containing the elements for which the function returned `True`.
- **Usage**: Typically used when you want to filter out unwanted elements from an iterable based on some condition.

#### Syntax:
filter(function, iterable)
```

- **`function`**: The function that tests each element of the iterable. If the function returns `True`,
 the element is included in the result; otherwise, it is excluded.
- **`iterable`**: The iterable whose elements will be filtered.

#### Example:
# Using filter to get even numbers
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))  # Output: [2, 4, 6]
```

- Here, `filter()` uses the `lambda x: x % 2 == 0` function to filter only the even numbers from the list.

---

### **Key Differences Between `map()`, `reduce()`, and `filter()`**

| **Function**  | **Purpose** | **Output** | **Usage** | **Example Use Case** |
|---------------|-------------|------------|-----------|----------------------|
| `map()`       | Applies a function to each item of the iterable | A transformed iterable | Used for applying a transformation to all elements
| Squaring each number in a list |
| `reduce()`    | Applies a function cumulatively to the items of the iterable | A single accumulated result
 | Used to reduce an iterable to a single value | Summing all numbers in a list |
| `filter()`    | Filters the iterable based on a condition (function returns `True` or `False`) |
 A filtered iterable | Used to select elements based on a condition | Getting all even numbers from a list |


### **Visualizing the Difference with an Example**

#### Example Scenario:
You have a list of numbers, and you want to:
- Square all the numbers (`map()`),
- Get the sum of the numbers (`reduce()`),
- Filter out the even numbers (`filter()`).

from functools import reduce

numbers = [1, 2, 3, 4, 5]

# map() to square each number
squared_numbers = map(lambda x: x ** 2, numbers)
print(list(squared_numbers))  # Output: [1, 4, 9, 16, 25]

# reduce() to get the sum of numbers
sum_of_numbers = reduce(lambda x, y: x + y, numbers)
print(sum_of_numbers)  # Output: 15

# filter() to get even numbers
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))  # Output: [2, 4]
```

### **Summary**
- **`map()`** applies a function to each item and returns a transformed iterable.
- **`reduce()`** applies a function cumulatively to reduce the iterable to a single value.
- **`filter()`** filters elements from an iterable based on a condition.
````

````python
 # Question 4 :  Write a Python function that checks if a given number is prime or not from 1 to 200.
 Answer : Here's a Python function that checks if a given number is prime, and it works for numbers from 1 to 200:

def is_prime(n):
    # Check if the number is less than 2 (not prime)
    if n < 2:
        return False

    # Check for factors from 2 to sqrt(n)
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False

    # If no factors were found, the number is prime
    return True

### Example usage:

# Check prime numbers from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(num)
```

### Explanation:
- **Prime numbers** are numbers greater than 1 that have no divisors other than 1 and themselves.
- The function first checks if the number is less than 2, returning `False` immediately
 since 1 and any negative number are not prime.
- Then, the function checks for factors from 2 to the square root of the number.
If any divisor is found, it returns `False`, indicating the number is not prime.
- If no divisors are found, the number is prime, and the function returns `True`.

This function can be used to check whether any number between 1 and 200 is prime.
````

````python
# Question 5 :  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.
Answer : Here’s a Python class that implements an iterator to generate the Fibonacci sequence up to a specified number of terms:

class FibonacciIterator:
    def __init__(self, n):
        # Initialize the number of terms in the Fibonacci sequence
        self.n = n
        self.a, self.b = 0, 1  # First two Fibonacci numbers
        self.count = 0  # Track the number of terms generated

    def __iter__(self):
        # The iterator object itself is returned
        return self

    def __next__(self):
        # Stop iteration if the count reaches the specified number of terms
        if self.count >= self.n:
            raise StopIteration

        # Generate the next Fibonacci number
        current = self.a
        self.a, self.b = self.b, self.a + self.b  # Update to the next two Fibonacci numbers
        self.count += 1

        return current
```

### Example usage:

# Create an iterator for the first 10 Fibonacci numbers
fib = FibonacciIterator(10)

# Iterate through the Fibonacci numbers
for num in fib:
    print(num)
```

### Explanation:
- The `FibonacciIterator` class implements the iterator protocol by defining the `__iter__()` and `__next__()` methods.
- **`__init__(self, n)`**: Initializes the iterator with `n` (the number of terms to generate),
 starting values of `a = 0` and `b = 1` (the first two Fibonacci numbers), and `count = 0` to keep track of how many terms have been generated.
- **`__iter__(self)`**: Returns the iterator object itself. This is necessary for the iterator protocol.
- **`__next__(self)`**: Generates the next Fibonacci number. It raises `StopIteration` once the sequence has produced `n` terms.

This iterator will allow you to loop through the Fibonacci sequence and stop once the specified number of terms is reached.
````

````python
# Question 9:  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
Answer . Here's a Python program that uses the `map()` function to convert a list of temperatures from Celsius to Fahrenheit:
### Code:

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temperatures = [0, 20, 37, 100, -5]

# Use map() to apply the conversion function to each element in the list
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Output the result
print(fahrenheit_temperatures)
```

### Output:

```
[32.0, 68.0, 98.6, 212.0, 23.0]
```

### Explanation:
1. **`celsius_to_fahrenheit(celsius)`**: This function takes a temperature in Celsius and returns the equivalent temperature in Fahrenheit.
   - The formula to convert Celsius to Fahrenheit is:
     \[
     \text{Fahrenheit} = \left(\text{Celsius} \times \frac{9}{5}\right) + 32
     \]

2. **`celsius_temperatures`**: This is a list of temperatures in Celsius that we want to convert.

3. **`map()`**: The `map()` function applies the `celsius_to_fahrenheit` function to each item
 in the `celsius_temperatures` list. The result of `map()` is an iterator, so we convert it to a list using `list()`.

4. **Output**: The list `fahrenheit_temperatures` contains the temperatures converted to Fahrenheit.
````

````python
# Question 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

Answer: To remove all the vowels from a given string using Python's `filter()` function, you can follow these steps:

1. Define a function that checks if a character is a vowel.
2. Use `filter()` to keep only the characters that are not vowels.
3. Convert the result back to a string.

Here's the Python code for this:

def remove_vowels(char):
    vowels = "aeiouAEIOU"
    return char not in vowels

# Input string
input_string = "Hello, this is a sample string."

# Use filter to remove vowels
result = ''.join(filter(remove_vowels, input_string))

print("Original string:", input_string)
print("String without vowels:", result)

### Explanation:
- `remove_vowels(char)` is a function that returns `True` if the character is not a vowel (i.e., it should be kept).
- `filter(remove_vowels, input_string)` filters out the vowels based on the function `remove_vowels`.
- `''.join()` is used to convert the filtered characters back into a string.

### Sample output:
```
Original string: Hello, this is a sample string.
String without vowels: Hll, ths s  smpl strng.

This code removes both uppercase and lowercase vowels from the given string.
````

````python
# Question 11 : ) Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:
#order Number      Book Title And Author             Quantity            Price Per item

35487            Learning Python, mark lutz            4                   40.95
98762            programming python, Mark Lutz         5                   56.80
77226            Head Python, Paul Barry               3                   32.95
88112            Einfuhrung in python3, Bernd Klein    3                   24.95

Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.

Write a Python program using lambda and map.
Answer : To solve this problem, we can follow these steps:

1. We'll first define a list of orders, where each order is represented as a sublist with the order number, book title, quantity, and price per item.
2. For each order, we'll calculate the total price by multiplying the quantity with the price per item.
3. If the total price is smaller than 100.00 €, we'll add 10.00 € to the value.
4. We will then return a list of 2-tuples, each containing the order number and the adjusted total price.
5. We'll use the `map()` function along with `lambda` to perform the required operations.

Here's how we can write the Python code:

# List of orders with sublists
orders = [
    [35487, 'Learning Python, Mark Lutz', 4, 40.95],
    [98762, 'Programming Python, Mark Lutz', 5, 56.80],
    [77226, 'Head Python, Paul Barry', 3, 32.95],
    [88112, 'Einfuhrung in Python3, Bernd Klein', 3, 24.95]
]

# Using lambda and map to calculate the order number and total price
result = list(map(lambda order: (
    order[0],  # order number
    (order[2] * order[3]) + 10 if order[2] * order[3] < 100 else order[2] * order[3]  # adjusted price
), orders))

# Print the result
print(result)
```

### Explanation:
- `orders`: This is the list containing sublists, where each sublist represents an order.
- `map()`: We use `map()` to apply a function to each item in the `orders` list.
- `lambda order`: The `lambda` function takes each order as input, and for each order:
  - We extract the order number (`order[0]`).
  - We calculate the total price (`order[2] * order[3]`), and if the value is less than 100, we add 10 to the total.
- `list()`: We convert the result of `map()` into a list.

### Output:
The output will be a list of 2-tuples, where each tuple contains the order number and the adjusted total price.

For the given input, the output will look like this:

```python
[
    (35487, 163.8),  # Total = 4 * 40.95 = 163.8 (no addition needed)
    (98762, 284.0),  # Total = 5 * 56.80 = 284.0 (no addition needed)
    (77226, 98.85),  # Total = 3 * 32.95 = 98.85 (+10 because it's less than 100)
    (88112, 89.85)   # Total = 3 * 24.95 = 74.85 (+10 because it's less than 100)
]






````
